# Retrieve Results for Simulation with LEO-based Communication Network

To run a simulation, execute script `main_agc_satnet.sh` from the root folder.
This notebook shows how to extract and visualize simulation results for the power system model.

Path to CSV file with results from power system model:

```python
from pathlib import Path
p_ps_out = Path('logs_ns3', 'agc_9busSystem_noSecondaryCtrl.csv').resolve(strict=True)
```

Import simulation results to dataframe:

```python
import pandas as pd
df_ps_out = pd.read_csv(p_ps_out)
df_ps_out.index = df_ps_out['time']
```

Plot frequency at substations:

```python
df_ps_out[['f1_bs', 'f2_bs', 'f3_bs']].plot()
```

Plot active power at substations:

```python
df_ps_out[['P1_bs', 'P2_bs', 'P3_bs']].plot()
```

Plot current at substations

```python
df_ps_out[['I1_bs', 'I2_bs', 'I3_bs']].plot()
```
