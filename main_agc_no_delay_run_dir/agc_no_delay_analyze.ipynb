# Retrieve Results for Simulation without Communication Delays

To run a simulation, execute script `main_agc_no_delay.sh` from the root folder.
This notebook shows how to extract and visualize simulation results for the power system model.

Path to CSV file with results from power system model:

```python
from pathlib import Path
p_ps_ref = Path('agc_9busSystem_out.csv').resolve(strict=True)
```

Import simulation results to dataframe:

```python
import pandas as pd
df_ps_ref = pd.read_csv(p_ps_ref)
df_ps_ref.index = df_ps_ref['time']
```

Plot frequency at substations:

```python
df_ps_ref[['f1_bs', 'f2_bs', 'f3_bs']].plot()
```

Plot active power at substations:

```python
df_ps_ref[['P1_bs', 'P2_bs', 'P3_bs']].plot()
```

Plot current at substations

```python
df_ps_ref[['I1_bs', 'I2_bs', 'I3_bs']].plot()
```
